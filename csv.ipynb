```python
import pandas as pd

# Define the testset name and paths for the input CSVs
testset_name = "ds_exact"
testset = "results/" + testset_name

properties_file = testset + "/properties.csv"
reductions = testset + "/reductions.csv"
findminhs_file = testset + "/findminhs.csv"
lp_file = testset + "/lp.csv"
#highs_file = testset + "/highs.csv"
#highs_reduced = testset + "/highs_reduced.csv"
scip_file = testset + "/scip.csv"
scip_reduced = testset + "/scip_reduced.csv"
domsat_file = testset + "/domsat.csv"
domsat_reduced = testset + "/domsat_reduced.csv"
nusc_file = testset + "/nusc.csv"
nusc_reduced = testset + "/nusc_reduced.csv"



# Read the CSV files into pandas DataFrames
df_properties = pd.read_csv(properties_file)
df_reductions = pd.read_csv(reductions)
df_findminhs = pd.read_csv(findminhs_file)
df_lp = pd.read_csv(lp_file)
#df_highs = pd.read_csv(highs_file)
#df_highs_reduced = pd.read_csv(highs_reduced)
df_scip = pd.read_csv(scip_file)
df_scip_reduced = pd.read_csv(scip_reduced)
df_domsat = pd.read_csv(domsat_file)
df_domsat_reduced = pd.read_csv(domsat_reduced)
df_nusc = pd.read_csv(nusc_file)
df_nusc_reduced = pd.read_csv(nusc_reduced)



# Merge the DataFrames on the 'Name' column
df_merged = pd.merge(df_properties, df_reductions, on="Name", how="left", suffixes=("", "_reductions"))
df_merged = pd.merge(df_merged, df_findminhs, on="Name", how="left", suffixes=("", "_findminhs"))
#df_merged = pd.merge(df_merged, df_highs, on="Name", how="left", suffixes=("", "_highs"))
#df_merged = pd.merge(df_merged, df_highs_reduced, on="Name", how="left", suffixes=("", "_highs_reduced"))
df_merged = pd.merge(df_merged, df_lp, on="Name", how="left", suffixes=("", "_lp"))
df_merged = pd.merge(df_merged, df_scip, on="Name", how="left", suffixes=("", "_scip"))
df_merged = pd.merge(df_merged, df_scip_reduced, on="Name", how="left", suffixes=("", "_scip_reduced"))
df_merged = pd.merge(df_merged, df_domsat, on="Name", how="left", suffixes=("", "_domsat"))
df_merged = pd.merge(df_merged, df_domsat_reduced, on="Name", how="left", suffixes=("", "_domsat_reduced"))
df_merged = pd.merge(df_merged, df_nusc, on="Name", how="left", suffixes=("", "_nusc"))
df_merged = pd.merge(df_merged, df_nusc_reduced, on="Name", how="left", suffixes=("", "_nusc_reduced"))





# Rename columns after the merge if necessary
"""
df_merged = df_merged.rename(columns={
    "Time Taken (seconds)": "Time FindMinHS [s]",
    "Solution Size": "Solution HiGHS",
    "Time Taken (seconds)_highs": "Time HiGHS [s]",
    "Solution Size_scip": "Solution SCIP",
    "Time Taken (seconds)_scip": "Time SCIP [s]",
    "Solution Size_lp": "LP Lower Bound",
    "Time Taken (seconds)_lp": "Time LP [s]",
    "Solution Size_domsat": "Solution DomSAT",
    "Time Taken (seconds)_domsat": "Time DomSAT [s]",
    "Solution Size_nusc": "Solution NuSC",
    "Time Taken (seconds)_nusc": "Time NuSC [s]"
})
"""

# Sort the merged DataFrame by the 'Name' column
df_merged = df_merged.sort_values(by="Name")

# Save the merged DataFrame to a new CSV file
df_merged.to_csv(testset + "/merged_output.csv", index=False)

# Display the merged DataFrame
print(df_merged)
```

```python
import pandas as pd

# Define the input and output CSV files
input_file = 'results/ds_exact/reductions.csv'
output_file = 'results/ds_exact/reductions_sorted.csv'

# Read the CSV file
df = pd.read_csv(input_file)  # replace with your actual file name

# Extract the number from the 'Name' column (e.g., "exact_001.gr" â†’ 1)
df["SortKey"] = df["Name"].str.extract(r"exact_(\d+)\.gr").astype(int)

# Sort by that number
df_sorted = df.sort_values(by="SortKey")

# Drop the helper column if you want
df_sorted = df_sorted.drop(columns=["SortKey"])

# Save to a new CSV (optional)
df_sorted.to_csv(output_file, index=False)

# Or print to console
print(df_sorted)
```
